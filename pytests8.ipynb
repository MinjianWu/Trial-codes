```python
# probability distribution function plot:
```

```python
np.random.seed(1234)
data = np.random.normal(0, 1, 10000)

sns.kdeplot(data, label='data')

plt.axvline(data.mean(), c='r', label='mean')
plt.axvline(data.mean() + data.std(), c='g', ls='--')
plt.axvline(data.mean() - data.std(), c='g', ls='--', label='one std')

plt.legend()

plt.show()
```

```python
# probability density function plot (Empirical Cumulative Density Function):
```

```python
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
plt.figure(figsize=(8,6))

ecdf = ECDF(data)
plt.plot(ecdf.x, ecdf.y, label='data')

plt.axhline(0.5, c='r')
plt.axvline(data.mean(), c='r', label='mean')

plt.axhline(0.025, c='y', ls='--')
plt.axvline(data.mean() - 2 * data.std(), c='y', ls='--')

plt.axhline(0.975, c='y', ls='--')
plt.axvline(data.mean() + 2 * data.std(), c='y', ls='--', label='two std - 95% conf')

plt.legend(loc='best')
plt.title('Probability Density Plot of Normal Distribution (μ=0, σ=1)')

plt.show()
```

```python
d = np.random.normal(62.5, 10, 100000)
x = sns.kdeplot(d)
plt.axvline(d.mean() + 1.96 * d.std())
```

```python
ecdf = ECDF(d)
plt.plot(ecdf.x, ecdf.y)

plt.axhline(0.975)
plt.axvline(d.mean() + 1.96 * d.std())
```
