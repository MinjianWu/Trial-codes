```python
# pivot tables are similar to groupby
# df.pivot_table parameters:
# values       - what values of interest, 
# index        - what to filter on, i.e. groupby what,
# aggfunc      - do what to the values in each group (default is np.mean)
# margins=True - show groups' overall aggfunc result
```

```python
pd.options.display.max_rows = 10        # this option to display both head(5) and tail(5)
happiness2015 = pd.read_csv("World_Happiness_2015.csv")
happiness2015
```

```python
# groupby method
happy_family_stats = happiness2015.groupby('Region')[['Happiness Score', 'Family']].agg([np.min, np.max, np.mean])
happy_family_stats
```

```python
# pivot table method
pv_happy_family_stats = happiness2015.pivot_table(
                                    values=['Happiness Score', 'Family'], 
                                    index=['Region'],
                                    aggfunc=[np.min, np.max, np.mean],
                                    margins=True)
pv_happy_family_stats
```

```python
import re
string = "JavaScript vs java"     # what if we just want 'java' (the second match)?
```

```python
pattern_1 = r"[jJ]ava"

m1 = re.findall(pattern_1, string)
print(m1)
```

```python
pattern_2 = r"[jJ]ava[^S]"

m2 = re.findall(pattern_2, string)
print(m2)
```

```python
pattern_3 = r"[jJ]ava[^S]?"

m3 = re.findall(pattern_3, string)
print(m3)
```

```python
pattern_4 = r"\b[jJ]ava\b"           # '\b' marks the word boundary - 'java' is a word

m4 = re.findall(pattern_4, string)
print(m4)
```

```python
# r'^[abc]'  - pattern any of 'a','b','c' only at the start of a string ('^' here is the beginning anchor)
# r'[^abc]'  - pattern NOT matching any of 'a','b','c' in a string ('^' here is the negative set indicator)
```

```python
email_tests = pd.Series(['email', 'Email', 'e Mail', 'e mail', 'E-mail',
              'e-mail', 'eMail', 'E-Mail', 'EMAIL', 'emails', 'Emails',
              'E-Mails', "late mail", "emailing"])

pattern_1 = r'e[\s-]?mails?'        
email_tests.str.contains(pattern_1, flags=re.I)
```

```python
# we only want variations of strings containing the likes of 'email'
# however, the above pattern captures "late mail" and "emailing", which is wrong!
# need to add word boundary anchors '\b'!
```

```python
pattern_2 = r'\be[\s-]?mails?\b'
email_tests.str.contains(pattern_2, flags=re.I)
```

```python
string1 = "I'm going to read a book"
re.search(r'(\w)\1', string1)             # (\w) is backreference 1 -> \1 will duplicate this reference 1
```

```python
string2 = "I'm going to read a long booooook"
re.search(r'(\w)\1{2}', string2)          # {2} will set the duplicate to repeat twice
```

```python
string3 = "I'm going to read a long booooook"
re.search(r'(\w)\1+', string3)            
```

```python
string4 = 'I say HelloGoodbyeGoodbyeHello and Hello and Goodbye'
re.search(r'(\w+)(\w+)\2\1', string4)     # each parenthesis contains a backreference group -> (1)(2)\2\1 pattern
```

```python
string5 = 'You say Hello and Hello back'
re.search(r'(\w+)\sand\s\1', string5)
```
