```python
fig = plt.figure()                # make a figure frame for subplots - 
                                  # all subplots must be generated and called in this same cell

ax1 = fig.add_subplot(2, 2, 1)    # figure size 2 x 2 subplots, ax1 resides top-left
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

plt.plot(np.random.randn(50).cumsum(), 'k--')     # default to execute on the LAST added subplot, i.e., ax3! 
                                                  # otherwise, call e.g., ax1.plot()

ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)   
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))  # overlay, not overwrite on subplot ax3

plt.plot(np.random.randn(50).cumsum(), 'r-')      # again, default (& overlay) to the LAST added subplot, ax3!
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'g--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_xlabel('stages')
ax.set_title('a plot')

ax.set_xlim([-100, 1100])
ax.set_ylim([-30, 30])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

props = {
    'xticks': [0, 250, 500, 750, 1000],
    'xticklabels': ['one', 'two', 'three', 'four', 'five'],
    'xlabel': 'stages',
    'title': 'a plot',
    'xlim': [-100, 1100],
    'ylim': [-30, 30]
}

ax.set(**props)     # alternative construction (batch set) using a dict
plt.show()
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')         # label parameter for legend
ax.plot(np.random.randn(1000).cumsum(), 'g--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'b', label='_nolegend')   # no label, no legend
ax.plot(np.random.randn(1000).cumsum(), 'y') 

ax.legend(loc='best')                                             # call .legend to add it to the plot
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
data = np.random.randn(1000).cumsum()
ax.plot(data, 'k')

ax.annotate('mark', xy=(300, data[300] + 5), 
            xytext=(300, data[300] + 15),
            arrowprops=dict(facecolor='red', headwidth=4, width=2, headlength=4),
            horizontalalignment='center', verticalalignment='top')
plt.show()
```

```python
plt.rc('figure', figsize=(8, 8))          # re-configure (customise) matplotlib plt styles
font_options = {'family': 'monospace',
                'weight': 'bold'}
plt.rc('font', **font_options)

'''---------------------------------------------------------------------------------------------------'''

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)   # (bottom-left coord), width, height
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)            # (center coord), radius
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.5)

ax.add_patch(rect)    # to add shapes
ax.add_patch(circ)
ax.add_patch(pgon)

ax.axis('equal')      # equalise x & y axes scales so a circle is a round shape
```
